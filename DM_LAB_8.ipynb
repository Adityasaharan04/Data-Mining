```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating a DataFrame with sales data over months
df = pd.DataFrame({'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
                   'Sales': [10000, 15000, 18000, 12000, 17000]})

# Plotting a line graph
plt.plot(df['Month'], df['Sales'], marker='o', linestyle='-', color='blue', label='Sales')
plt.xlabel('Month')
plt.ylabel('Sales in INR')
plt.title('Monthly Sales Trend')
plt.legend()
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# First subplot: Line Plot
ax[0].plot(df['Month'], df['Sales'], marker='o', color='red')
ax[0].set_title('Sales Trend')
ax[0].set_xlabel('Month')
ax[0].set_ylabel('Sales in INR')

# Second subplot: Bar Plot
ax[1].bar(df['Month'], df['Sales'], color='green')
ax[1].set_title('Sales Bar Plot')
ax[1].set_xlabel('Month')
ax[1].set_ylabel('Sales in INR')

plt.tight_layout()
plt.show()
```

```python
plt.bar(df['Month'], df['Sales'], color='purple')
plt.xlabel('Month')
plt.ylabel('Sales in INR')
plt.title('Monthly Sales Comparison')
plt.show()
```

```python
import numpy as np

# Generating random sales data
sales_data = np.random.normal(15000, 3000, 100)

# Plotting histogram
plt.hist(sales_data, bins=10, color='orange', edgecolor='black')
plt.xlabel('Sales in INR')
plt.ylabel('Frequency')
plt.title('Sales Distribution')
plt.show()
```

```python
# Creating a DataFrame for product sales
df_pie = pd.DataFrame({'Product': ['Laptop', 'Phone', 'Tablet', 'TV'],
                        'Sales': [30000, 50000, 20000, 40000]})

# Plotting pie chart
plt.pie(df_pie['Sales'], labels=df_pie['Product'], autopct='%1.1f%%', startangle=90, colors=['blue', 'green', 'red', 'purple'])
plt.title('Product Sales Distribution')
plt.show()
```

```python
# Creating random sales vs. profit data
profit = [3000, 4000, 5000, 2000, 6000]

# Plotting scatter plot
plt.scatter(df['Sales'], profit, color='brown')
plt.xlabel('Sales in INR')
plt.ylabel('Profit in INR')
plt.title('Sales vs. Profit')
plt.show()
```

```python
# Creating a box plot
plt.boxplot(df['Sales'], vert=False, patch_artist=True, boxprops=dict(facecolor='cyan'))
plt.xlabel('Sales in INR')
plt.title('Sales Data Distribution')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Setting a theme for better visualization
sns.set_theme(style="darkgrid")

# Creating sample data
df = sns.load_dataset("penguins")

# Plotting scatter plot with aesthetic settings
plt.figure(figsize=(8, 5))
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm", hue="species", data=df, palette="coolwarm")
plt.title("Penguin Bill Length vs Bill Depth")
plt.show()
```

```python
# Setting a different color palette
sns.set_palette("pastel")

# Plotting bar plot with custom color palette
plt.figure(figsize=(8, 5))
sns.barplot(x="species", y="body_mass_g", data=df, palette="Blues")
plt.title("Average Body Mass of Penguin Species")
plt.show()

```

```python
# Plotting histogram for bill length
plt.figure(figsize=(8, 5))
sns.histplot(df["bill_length_mm"], bins=20, kde=True, color="purple")
plt.title("Distribution of Bill Length in Penguins")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Creating a pairplot for numerical columns
sns.pairplot(df, hue="species", palette="husl")
plt.show()
```

```python
# Box plot to compare flipper length across species
plt.figure(figsize=(8, 5))
sns.boxplot(x="species", y="flipper_length_mm", data=df, palette="Set2")
plt.title("Comparison of Flipper Length Among Species")
plt.show()
```

```python
# KDE Plot to visualize distribution of flipper length
plt.figure(figsize=(8, 5))
sns.kdeplot(df["flipper_length_mm"], shade=True, color="green")
plt.title("Density Plot for Flipper Length")
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Density")
plt.show()
```
