```python
import matplotlib.pyplot as plt

# Given data
age = [23, 23, 27, 27, 39, 41, 47, 47, 49, 50, 52, 54, 56, 57, 58, 58, 60, 61]
fat_percentage = [9.5, 26.5, 7.8, 17.8, 31.4, 25.9, 27.4, 27.2, 31.2, 34.6, 42.5, 28.8, 33.4, 30.2, 34.1, 32.9, 41.2, 35.7]

# Creating the boxplot
plt.figure(figsize=(10,5))

# Subplot 1: Boxplot for Age
plt.subplot(1,2,1)
plt.boxplot(age, vert=True, patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.title('Boxplot of Age')
plt.ylabel('Age')

# Subplot 2: Boxplot for %Fat
plt.subplot(1,2,2)
plt.boxplot(fat_percentage, vert=True, patch_artist=True, boxprops=dict(facecolor="lightgreen"))
plt.title('Boxplot of %Fat')
plt.ylabel('%Fat')

# Show the plots
plt.tight_layout()
plt.show()
```

```python
from graphviz import Digraph

# Create a directed graph
dot = Digraph(format='png')

# Fact Table
dot.node('Fact', 'Student Performance Fact\n(Student_ID, Course_ID, Semester_ID, Instructor_ID, Count, Avg Grade)', shape='box', style='filled', fillcolor='lightblue')

# Dimension Tables
dot.node('Student', 'Student Dimension\n(Student_ID, Name, Dept_ID)', shape='box', style='filled', fillcolor='lightgray')
dot.node('Course', 'Course Dimension\n(Course_ID, Subject_ID, Course Name)', shape='box', style='filled', fillcolor='lightgray')
dot.node('Semester', 'Semester Dimension\n(Semester_ID, Year, Term)', shape='box', style='filled', fillcolor='lightgray')
dot.node('Instructor', 'Instructor Dimension\n(Instructor_ID, Name, Dept_ID)', shape='box', style='filled', fillcolor='lightgray')

# Normalized Tables (Snowflake structure)
dot.node('Department', 'Department Table\n(Dept_ID, Dept_Name)', shape='box', style='filled', fillcolor='white')
dot.node('Subject', 'Subject Table\n(Subject_ID, Subject Name)', shape='box', style='filled', fillcolor='white')

# Connecting Fact Table to Dimension Tables
dot.edge('Fact', 'Student')
dot.edge('Fact', 'Course')
dot.edge('Fact', 'Semester')
dot.edge('Fact', 'Instructor')

# Connecting Dimension Tables to Normalized Tables
dot.edge('Student', 'Department', label='Dept_ID')
dot.edge('Instructor', 'Department', label='Dept_ID')
dot.edge('Course', 'Subject', label='Subject_ID')

# Render and display the graph
dot.render('snowflake_schema', format='png', view=True)
```

```python
import matplotlib.pyplot as plt

# Time points and corresponding number of customers in the system L(t)
time_points = [0.4, 1.6, 2.1, 2.4, 3.1, 3.8, 4.0, 5.1, 5.6, 6.2, 8.6, 9.9, 10.5]
num_customers = [1, 2, 3, 2, 1, 2, 3, 2, 3, 2, 1, 0, 0]  # L(t)

# Plot the step function
plt.step(time_points, num_customers, where='post', linestyle='-', marker='o', color='b', label="L(t)")

# Labels and title
plt.xlabel("Time (minutes)")
plt.ylabel("Number of Customers in System L(t)")
plt.title("Step Plot of L(t) vs. Time")
plt.grid(True, linestyle="--", alpha=0.7)
plt.legend()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Time intervals for Fixed-Increment (t = 1 unit)
fixed_time = np.arange(0, 11, 1)  # Fixed steps of 1
fixed_values = np.ones_like(fixed_time)  # Just to show equal spacing

# Time intervals for Next-Event (Random event times)
event_times = [0, 2.4, 5.8, 7.2, 10]  # Event-based jumps
event_values = np.ones_like(event_times) * 2  # Different row to differentiate

plt.figure(figsize=(8, 4))
plt.scatter(fixed_time, fixed_values, color='blue', label="Fixed-Increment Time Advance", marker='o')
plt.scatter(event_times, event_values, color='red', label="Next-Event Time Advance", marker='o')

# Labels and styling
plt.xlabel("Time")
plt.ylabel("Simulation Step")
plt.title("Comparison of Next-Event and Fixed-Increment Time Advance")
plt.yticks([1, 2], ["Fixed-Increment", "Next-Event"])
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()

```
